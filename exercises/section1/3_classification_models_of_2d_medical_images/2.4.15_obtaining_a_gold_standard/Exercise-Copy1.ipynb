```python
import numpy as np
import pandas as pd
```

Read in your label data:

```python
labels = pd.read_csv('labels.csv')
labels.head(10)
```

## Create your first ground truth as derived from biopsy labels: 

```python
bx_gold = []
for i, r in labels.iterrows():
    if r['biopsy'] == 'benign': bx_gold.append(0)
    else: bx_gold.append(1)
        
labels['bx_gold'] = bx_gold
labels.head()
```

## Create your second truth by voting system from the three radiologists:

```python
rad_silver = []
for i, r in labels.iterrows():
    rads = [r['rad1'], r['rad2'], r['rad3']]
    selection = max(set(rads), key = rads.count)
    if selection == 'benign': rad_silver.append(0)
    else: rad_silver.append(1)    

labels['rad_silver']=rad_silver
labels.head()
    
```

## Create your third ground truth by weighting the three radiologists:

```python
rad_silver_w = []
for i, r in labels.iterrows():
    rads = [r['rad1'], r['rad2'], r['rad3']]
    rads_w = []
    for x in rads:
        if rads.index(x) == 0:
            w = 5/30
        elif rads.index(x) == 1:
            w = 10/30
        else:
            w = 0.5
            
        if x  == 'benign': x=0*w
        else: x = 1*w
        rads_w.append(x)
    
            
    if sum(rads_w) >= 0.5 :
        selection = 1
    else:
        selection = 0
    rad_silver_w.append(selection)

labels['rad_silver_w']=rad_silver_w
labels
```

## Compare the three ground truths:

Here, just explore the three sets of labels you created and see how often they agree

```python
len(labels[(labels['bx_gold'] == labels['rad_silver'])])/len(labels)
```

```python
len(labels[(labels['bx_gold'] == labels['rad_silver_w'])])/len(labels)
```
